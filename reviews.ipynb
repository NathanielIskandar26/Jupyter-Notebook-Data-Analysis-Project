### Aggregate Data (Grouping), Downsampling, and Plotting

##### Aggregate Data by Day and Plotting it

```python
df['Day'] = df['Timestamp'].dt.date
day_average = df.groupby(['Day']).mean('Rating')

plt.figure(figsize=(25, 5))
plt.plot(day_average.index, day_average['Rating'])
```

##### Aggregate Data by Week, Downsampling, and Plotting it

```python
df['Week'] = df['Timestamp'].dt.strftime('%Y-%U') #Downsample
week_average = df.groupby(['Week']).mean('Rating')

plt.figure(figsize=(20, 5))
plt.plot(week_average.index, week_average['Rating'])
```

##### Aggregate Data by Month, Downsampling, and Plotting it

```python
df['Month'] = df['Timestamp'].dt.strftime('%Y-%m') #Downsample
month_average = df.groupby((['Month'])).mean('Rating')

plt.figure(figsize=(50, 5))
plt.plot(month_average.index, month_average['Rating'])
```

##### Average Rating by Month by Course

```python
df['Month'] = df['Timestamp'].dt.strftime('%Y-%m') #Downsample
month_average_crs = df.groupby(['Month', 'Course Name'])['Rating'].mean('Rating').unstack()

month_average_crs.plot(figsize=(25, 8))
```

##### Total Rating by Month by Course

```python
df['Month'] = df['Timestamp'].dt.strftime('%Y-%m') #Downsample
month_average_crs = df.groupby(['Month', 'Course Name'])['Rating'].count().unstack()

month_average_crs.plot(figsize=(25, 8))
```

##### EX Case 1: What Day Do People Tend to Give a More Positive Review?

```python
df['Weekday'] = df['Timestamp'].dt.strftime('%A')
df['Daynumber'] = df['Timestamp'].dt.strftime('%w')

weekday_average = df.groupby(['Weekday', 'Daynumber']).mean('Rating')
weekday_average = weekday_average.sort_values('Daynumber')

plt.plot(weekday_average.index.get_level_values(0), weekday_average['Rating'])
```
